# Distribuci√≥n Horseshoe

```python
import arviz as az
import pymc as pm
import preliz as pz
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
```

Vamos a intentar simular distribuciones del prior horseshoe a partir de las distribuciones que lo conforman.

```python
# Simulo observaciones del parametro lambda de una HalfCauchy
cauchy1 = pz.Cauchy(0,1).rvs(size=3000, random_state=None)
param1 = abs(cauchy1)

# Creo un gr√°fico para ver la distribuci√≥n de Lambda
plt.hist(param1, bins=100, edgecolor='black')
plt.title('Histograma de Distribuci√≥n Half-Cauchy')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Mostrar el histograma
plt.show()

# Defino los par√°metros que faltan, y realizo la simulaci√≥n de los datos de 
tau = 0.01 # Este podr√≠a ser un hiperpar√°metro si se desea

beta = pz.Normal(0, sigma=tau*param1).rvs(3000)

# Creo un gr√°fico para ver la distribuci√≥n de Beta
plt.hist(beta, bins=100, edgecolor='black')
plt.title('Histograma del prior de Beta simulado')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.ylim(top=200)
# Mostrar el histograma
plt.show()
```
